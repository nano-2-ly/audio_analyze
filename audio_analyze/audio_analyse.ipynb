```python
import librosa
import librosa.display
import IPython.display
import numpy as np
import matplotlib.pyplot as plt


audio_path = "./test_data_set/3gp.3gp"
y, sr = librosa.load(audio_path)

IPython.display.Audio(data=y, rate=sr)

#np.set_printoptions(threshold=np.nan)
```

```python
print(y.shape)
plt.plot(y)
plt.show()
```

```python
audio_path = "./test_data_set/mp3.mp3"
z, sr = librosa.load(audio_path)

IPython.display.Audio(data=z, rate=sr)
```

```python
print(z.shape)
plt.plot(z)
plt.show()

print(z)
```

```python
audio_path = "./test_data_set/m4a.m4a"
x, sr = librosa.load(audio_path)

print(x.shape)
plt.plot(x)
plt.show()

print(x)
```

```python
audio_path = "./test_data_set/wav.wav"
w, sr = librosa.load(audio_path)

print(w.shape)
plt.plot(w)
plt.show()

print(w)
```

```python
start = 24950+100
end = 25000+100

plt.plot(x[start:end],'r', label = "m4a")
plt.plot(y[start:end],'b', label = "3gp")
plt.plot(z[start+844: end+844],'g', label = "mp3")
plt.plot(w[start:end],'y', label = "wav")

#print(x[start:end])

#print(y[start:end])

#print(w[start:end])

plt.legend(loc='upper right')
plt.show()
```
